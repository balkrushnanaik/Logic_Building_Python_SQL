31. Write a Python program to print numbers from 1 to 10 using a loop.


```python
n = 10
for i in range(1,n+1):
    print(i)
```

32. Write a Python program to print even numbers from 1 to 100.


```python
n = 100
for i in range(1,n+1):
    if i % 2 == 0:
        print(i)
```

33. Write a Python program to print odd numbers from 1 to 100.


```python
n = 100
for i in range(1,n+1):
    if i % 2 != 0:
        print(i)
```

35. Write a Python program to calculate the factorial of a number.
